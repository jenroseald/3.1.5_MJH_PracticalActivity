# Combine using each of the 3 methods

### Combine using Merge

```python
# Shared column is ID
# Left DF is Movies, Right DF is Ott
```

```python
movies_ott_merg = pd.merge(movies, ott, how='left', on='ID')
```

```python
# View new DF
movies_ott_merg
```

### Combine using Concat

```python
movies_ott_conc = pd.concat([movies, ott], axis=0)
```
